```python
import matplotlib.pyplot as plt
import numpy as np
import jax
import jax.numpy as jnp
import jax.nn as jnn
import jax.random as jr
from jaxtyping import Float, Array, Key
import diffrax
import equinox as eqx
import pandas as pd
import optax
import time

from models.SingleSolveCDE import SingleSolveCDE
from models.DataLoader import DataLoader
from utils import train_utils, plots
```

```python

best_test_mse = float('inf')
test_mses = []

#need to store this since all __init__ parameters are required for deserialization
best_seed = None


for seed in [0, 1, 2, 3, 4]:
    key = jr.key(seed)
    model_key, dataloader_key = jr.split(key, 2)

    model = SingleSolveCDE(data_size=3, hidden_size=3, width_size=64, depth=3, key=model_key)
    dataloader = DataLoader("data/data_large.npy", key=dataloader_key)

    start_time = time.time()
    model = train_utils.train_until_day_100(model, dataloader)
    end_time = time.time()

    print(f"Trained model in {end_time-start_time:.2f}")

    test_loss = train_utils.test_loss_until_100(model, dataloader.test_data)

    if test_loss < best_test_mse:
        best_test_mse = test_loss
        best_seed = seed
        eqx.tree_serialise_leaves("serialised_models/single_solve_cde_long.eqx", model)

    test_mses.append(test_loss)

print(f"Best seed: {best_seed}, best test mse: {best_test_mse}")
print(f"Mean test MSE: {jnp.mean(jnp.array(test_mses))}")
print(f"Standard deviation test MSE: {jnp.std(jnp.array(test_mses))}")

```

```python
fname = 'data/clipped_data.csv'
df = pd.read_csv(fname)
real_world_ys = jnp.array(df['new_cases_7d_average'].values) 
ys = jnp.log10(1 + real_world_ys)

model = eqx.tree_deserialise_leaves("serialised_models/single_solve_cde_long.eqx", like=SingleSolveCDE(data_size=3, hidden_size=3, width_size=64, depth=3, key=model_key))


def plot_predictions(model: eqx.Module, ys: jnp.ndarray, control_until: int):
    """
        The code in this method was taken from
        my advisor Umur Can Kaya. No reference
        since the code is in a private repository.

        Plots the ground truth and prediction of 
        a single time series as well as a vertical
        line to indicate when the input ends.
    """
    ts = jnp.linspace(0, 1, 100) 

    plot_until = control_until + 14

    control_ys = ys[:control_until]
    control_ys, mean, std = train_utils.standardize(control_ys)
    y_pred = model(ts=ts, ys=control_ys, control_until=control_until, saveat=ts, train_until=100)
    y_pred = train_utils.inverse_log_transform(train_utils.destandardize(y_pred, mean, std))

    fig, ax = plt.subplots(1, 1,figsize=(5,5))


    
    ax.plot(ts, y_pred, label='Fit')
    ax.scatter(ts, train_utils.inverse_log_transform(ys),label='Data', marker='.', ls='', c='k')
    ax.set(xlabel=r'$t$ (days)', ylabel=r'$y$', yscale='linear')
    ax.grid(True)
    ax.legend(loc='upper right')

    ax.axvline(ts[control_until-1], c='k', ls='--', label='control until')

    plt.tight_layout()
    plt.show()


def plot_predictions_with_different_input_lengths(model, ys):
    ys = jnp.log10(1 + ys)
    ts = jnp.linspace(0, 1, 100)

    for control_until in [10, 20, 30, 40, 50, 60, 70, 80, 90]:
        plot_predictions(model, ys, control_until=control_until)

plot_predictions_with_different_input_lengths(model, real_world_ys)
```
